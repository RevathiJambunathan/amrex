```python
import h5py
import numpy as np
import yt
```

```python
print(f['level_0']['data:datatype=0'])
```

```python
d = np.reshape(f['level_0']['data:datatype=0'], (128, 128, 128, 1), order='F')
```

```python
d0 = d[:,:,:,0]
```

```python
data = {"component0":(d0,"cm")}
bbox = np.array([[-0.5, 0.5], [-0.5, 0.5], [-0.5, 0.5]])
```

```python
ds = yt.load_uniform_grid(data, np.shape(d0), length_unit=1.0, bbox=bbox, nprocs=8, 
                          periodicity=(False,False,False))
```

```python
ds.field_list
```

```python
prj = yt.ProjectionPlot(ds, "z", ["component0"])
prj.set_log("component0", False)
prj.show()
```
